# E-Commerce Data Analysis - October 2019
Analyzing the 2019-Oct.csv dataset to understand e-commerce trends and patterns.

## 1. Import Required Libraries
Import pandas, NumPy, and other necessary libraries for data manipulation and analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
print("Libraries imported successfully!")
```

## 2. Set Data Directory Path
Define the path to the data directory and specify the CSV file name (2019-Oct.csv).

```python
import os
from pathlib import Path

# Set data directory path
# In Jupyter, use current working directory instead of __file__
data_dir = os.getcwd()
csv_file = '2019-Oct.csv'
file_path = os.path.join(data_dir, csv_file)

print(f"Data directory: {data_dir}")
print(f"CSV file: {csv_file}")
print(f"Full path: {file_path}")
print(f"File exists: {os.path.exists(file_path)}")
```

## 3. Load CSV File
Use pandas read_csv() function to load the 2019-Oct.csv file into a DataFrame.

```python
# Load the CSV file
df = pd.read_csv(file_path)

print(f"✓ Data loaded successfully!")
print(f"Dataset shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

## 4. Display Data Preview
Display the first few rows of the DataFrame to preview the data.

```python
# Display first few rows
print("First 5 rows of the dataset:")
print(df.head())
```

## 5. Check Data Structure
Examine the data structure using info(), shape, and describe() methods to understand the dataset.

```python
# Display data information
print("=" * 80)
print("DATA INFORMATION")
print("=" * 80)
print(f"\nDataset Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\nColumn Names and Types:")
print(df.dtypes)

print("\n" + "=" * 80)
print("DATA SUMMARY STATISTICS")
print("=" * 80)
print(df.describe())

print("\n" + "=" * 80)
print("MISSING VALUES")
print("=" * 80)
print(df.isnull().sum())
```
